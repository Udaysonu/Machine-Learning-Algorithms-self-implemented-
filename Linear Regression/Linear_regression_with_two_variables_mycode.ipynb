```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("./data.csv",delimiter=",")
```

```python
x_train=data[:,0]
y_train=data[:,1]

#class which do the work of linear regression
class Linear_regression:
    #init takes number_of_iterations and learning_rate but if you not pass 
    #then by default learning rate is 0.0001 and no_of_iterations is 100
    def __init__(self,ln=0.0001,it=100):
        self.m=0
        self.c=0
        self.learning_rate=ln
        self.iterations=it
    
    #function to perform gradient descent
    def gd(self,x,y,m,c):
        dho_m=0
        dho_c=0
        
        #(dCost/dm)=sigma of (-2/no of data values)*(y_actual-y_predited)*m
        #(dCost/dc)=sigma of (-2/no of data values)*(y_actual-y_predicted)
        
        for i in range(len(x)):
            dho_m+=(-2/len(x))*(y[i]-m*x[i]-c)*x[i]
            dho_c+=(-2/len(x))*(y[i]-m*x[i]-c)
        
        # m=m-(alpha)*(dy/dm)
        #c=c-(alpha)*(dy/dc)
        #alpha=learning Rate
        return m-self.learning_rate*dho_m,c-self.learning_rate*dho_c
    
    #funciton to 
    def fit(self,x_train,y_train):
        
        #initializing the values of m,c (as our wish)
        m,c=0,0
        
        #running iterations 
        for i in range(self.iterations):
            m,c=self.gd(x_train,y_train,m,c)
            
            #Plotting the graph
            plt.plot(x_train,m*x_train-c,color="orange")
            plt.scatter(x_train,y_train)
            
            #showing iteration number and cost of function
            print("Iteration Number",i+1,"---Cost----",self.cost(x_train,y_train,m,c))

            plt.show()
        #storing the values of m,c in the class instance
        self.m=m
        self.c=c
    #function calculates the cost of our prediction
    def cost(self,x,y,m,c):
        return sum(y-m*x-c)/len(x)
    #funciton returns coefficient calculated 
    def coef(self):
        return self.m
    #function returns intercept calculated
    def intercept(self):
        return self.c
lr=Linear_regression()
lr.fit(x_train,y_train)
```

```python
print("Coefficient",lr.coef())
print("Intercept",lr.intercept())

#code written Uday Kiran Bakka :)
```
